```python
import pandas as pd
```

```python
WorldCupMatches = 'WorldCupMatches.csv'
WorldCupPlayers = 'WorldCupPlayers.csv'
WorldCups = 'WorldCups.csv'
```

```python
df_mat = pd.read_csv(WorldCupMatches)
df_pla = pd.read_csv(WorldCupPlayers)
df_wc = pd.read_csv(WorldCups)
```

```python
#Check percentage of empty cells per column
for x in df_mat.columns:
    print(f"The percentage of empty cells in column {x} is {round(df_mat[x].isna().sum()/len(df_mat[x])*100,2)} %.")
```

```python
#Check percentage of empty cells per column
for x in df_pla.columns:
    print(f"The percentage of empty cells in column {x} is {round(df_pla[x].isna().sum()/len(df_pla[x])*100,2)} %.")
```
